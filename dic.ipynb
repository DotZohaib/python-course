
## Python mein **Dictionary** (jise short mein 'dict' kehte hain) sabse zyada use hone wala aur powerful data structure hai.

### Bilkul aasan alfazon mein: Jaise aapki real-life phonebook mein ek **"Naam" (Key)** hota hai aur uske aage uska **"Phone Number" (Value)** likha hota hai, Python dictionary bilkul waise hi kaam karti hai. Isme data hamesha **Key: Value** ki jori (pair) mein save hota hai.


## Step 1: Dictionary Kaise Banate Hain? (The Key-Value Format)

Dictionary banane ke liye bhi Curly Braces `{}` use hote hain (jaise Set mein thay), lekin isme har item ke do hissay hote hain, jinhein colon `:` se alag kiya jata hai.


```python
# 1. Normal Dictionary
student = {
    "name": "Zohaib",
    "age": 22,
    "course": "Python"
}
print(type(student)) 
# Output: <class 'dict'>

# 2. Khali (Empty) Dictionary bananan
# Yaad rakhein, khali {} hamesha Dictionary banta hai, Set nahi!
empty_dict = {}
print(type(empty_dict)) 
# Output: <class 'dict'>
```

## Step 2: Sabse Bada Rule — Keys hamesha Unique hoti hain!

Dictionary ki **Key** kabhi repeat nahi ho sakti. Agar aap ghalti se ek hi Key do dafa likh dein, toh Python error nahi dega, balkay **purani value ko nayi value se replace (overwrite) kar dega**.


```python
car = {
    "brand": "Toyota",
    "model": "Corolla",
    "year": 2015,
    "year": 2023  # Same key dobara use ki
}

print(car)
# Output: {'brand': 'Toyota', 'model': 'Corolla', 'year': 2023} 
# (Dekhein, 2015 replace ho kar 2023 ban gaya)
```


> **Interview Tip:** Keys hamesha "Immutable" (jo change na ho sake) data types honi chahiye, jaise Strings, Numbers, ya Tuples. Aap kisi List ko Dictionary ki Key nahi bana sakte.



## Step 3: Data Nikalne Ka Tarika (The `.get()` Trap)

Agar humein dictionary se koi value nikalni hai, toh hum Index (`[0], [1]`) use nahi karte, balkay uski **Key** ka naam likhte hain. Lekin yahan beginners ek bohot bari ghalti karte hain.


```python
user = {"username": "ali_123", "email": "ali@gmail.com"}

# 1. Normal tareeqa (Bracket Notation)
print(user["username"]) # Output: ali_123

# GHALAT tareeqa: Agar Key exist nahi karti, toh code CRASH ho jayega!
# print(user["password"]) 
# KeyError: 'password'

# 2. SAHI aur Safe tareeqa (The .get() method)
# Agar key nahi hogi, toh yeh Error nahi dega balkay 'None' dega.
print(user.get("password")) 
# Output: None

# Aap chahain toh 'None' ki jagah apna custom message bhi de sakte hain:
print(user.get("password", "Bhai password nahi mila!"))
# Output: Bhai password nahi mila!
```

## Step 4: Data Add, Update, aur Remove Kaise Karein?

Dictionary bohot flexible hoti hai. Isme naya data dalna ya purana delete karna bohot aasan hai.


```python
profile = {"name": "Sara", "age": 25}

# 1. Naya data Add karna
profile["city"] = "Karachi"

# 2. Purana data Update karna
profile["age"] = 26 

# 3. Update method (Ek sath zyada keys add/update karni hon)
profile.update({"country": "Pakistan", "role": "Developer"})

# 4. Data Remove karna (.pop() method)
# pop() woh value return bhi karta hai jo usne delete ki hai
removed_age = profile.pop("age") 

print(profile)
# Output: {'name': 'Sara', 'city': 'Karachi', 'country': 'Pakistan', 'role': 'Developer'}
```

## Step 5: Loops aur Dictionary (Asal Taqat)

Jab aapke paas bohot sara data ho, toh usko handle karne ke liye hum loops lagate hain. Dictionary mein loop lagane ke 3 main methods hain:


```python
student_marks = {"Maths": 90, "Science": 85, "English": 78}

# 1. Sirf Keys nikalne ke liye:
for subject in student_marks.keys():
    print(subject)
# Output: Maths, Science, English

# 2. Sirf Values nikalne ke liye:
for marks in student_marks.values():
    print(marks)
# Output: 90, 85, 78

# 3. Dono (Keys aur Values) ek sath nikalne ke liye (Sabse zyada use hota hai):
for subject, marks in student_marks.items():
    print(f"{subject} mein marks hain: {marks}")
# Output: 
# Maths mein marks hain: 90
# Science mein marks hain: 85
# English mein marks hain: 78
```
