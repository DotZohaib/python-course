## 1. Jinki Memory Hamesha SAME Hoti Hai (Cached / Singletons)

Python in cheezon ko program start hone par hi memory mein bana leta hai. Jab bhi aap inko use karte hain, Python hamesha purana address hi deta hai:

* **Booleans (`True` aur `False`):** Python mein `True` aur `False` sirf ek hi baar bante hain.

```python
x = True
y = True
# id(x) aur id(y) hamesha same honge
print(id(x)) 
print(id(y))
```


* **NoneType (`None`):** `None` ka memory address hamesha same rehta hai.
* **Chotay Numbers (Small Integers):** Jaisa humne pehle discuss kiya, `-5` se lekar `256` tak.
* **Empty Tuples `()`:** Kyunki tuples change nahi ho sakte (immutable hote hain), isliye khali tuple hamesha ek hi memory share karta hai.

## 2. Jinki Memory Hamesha ALAG Hoti Hai (No Caching)

Ye wo data types hain jinke liye Python hamesha naya box (memory location) banata hai, chahe unki value same hi kyun na ho:

* **Bare Integers:** `-6` se chotay ya `257` se bare numbers. (Note: Agar aap ek hi file mein code likh rahe hain toh Python optimization ki wajah se same de sakta hai, lekin terminal par hamesha alag dega).

```python
a = 257
b = 257
# id(a) aur id(b) alag honge
print(id(a)) 
print(id(b))
```


* **Floats (Decimal Numbers):** Python floats ko cache nahi karta.

```python
a = 3.14
b = 3.14
# id(a) aur id(b) alag honge
print(id(a)) 
print(id(b))
```

* **Mutable Data Types (Lists, Dictionaries, Sets):** Ye sabse important rule hai! Jo cheezein baad mein change (modify) ho sakti hain, Python unko **kabhi** same memory nahi deta.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]
# Inka id() HAMESHA alag hoga
print(id(list1)) 
print(id(list2))
```

*Logic:* Agar Python `list1` aur `list2` ko same memory de deta, toh aap `list1` mein kuch add karte, toh wo automatically `list2` mein bhi add ho jata! Is confusion se bachne ke liye inki memory hamesha alag hoti hai.

## Quick Summary Table

| Data Type | Example | Memory Address (ID) |
| --- | --- | --- |
| **Booleans & None** | `True`, `False`, `None` | Hamesha SAME |
| **Small Integers** | `-5` se `256` | Hamesha SAME |
| **Simple Strings** | `"hello"`, `"python"` | Aksar SAME (Interned) |
| **Empty Tuples** | `()` | Hamesha SAME |
| **Large Integers** | `257`, `1000`, `-10` | Hamesha ALAG |
| **Floats** | `1.5`, `9.99` | Hamesha ALAG |
| **Lists, Dicts, Sets** | `[]`, `{}`, `[1, 2]` | Hamesha ALAG |
