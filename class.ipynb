
## 1. Core Concept: Class vs. Object


| Term | Aam Zindagi Ki Example | Programming Ki Zuban Mein |
| --- | --- | --- |
| **Class** | Ghar ka Naksha (Blueprint) | Ek template ya structure jisme variables aur functions define hote hain. |
| **Object** | Us nakshe par bana hua actual Ghar | Class ka ek real instance jo memory space leta hai. |

Ek nakshe (Class) se aap hazaron ghar (Objects) bana sakte hain. Har ghar ka color ya size alag ho sakta hai, par basic structure nakshe wala hi rahega.

---

## 2. Basic Syntax & `__init__` (The Constructor)

Jab aap code likhein, toh sab se pehle ek simple structure dikhayein:

```python
class Student:
    # Constructor: Jab bhi naya student banega, yeh khud chalega
    def __init__(self, name, roll_number):
        self.name = name          # Attribute
        self.roll_number = roll_number
    
    # Method (Class ke andar function ko method kehte hain)
    def introduce(self):
        print(f"Hello, my name is {self.name} and my roll number is {self.roll_number}")

# Objects banana (Instantiation)
student1 = Student("Zohaib", "101")
student2 = Student("Ali", "102")

student1.introduce() # Output: Hello, my name is Zohaib...
```


### Yahan 3 cheezein samjhana zaroori hain:

1. **`__init__` kya hai?** Yeh ek special "magic" method hai (jise dunder method bhi kehte hain). Jaise hi aap `Student("Zohaib", "101")` likhte hain, Python automatically `__init__` ko call karta hai taake naye object ko initial data de sake.
2. **Attributes vs Variables:** Class ke andar jo variables hote hain (jaise `name`, `roll_number`), unhein **Attributes** kaha jata hai.
3. **Methods vs Functions:** Class ke andar jo functions hote hain (jaise `introduce`), unhein **Methods** kaha jata hai.

## 3. The Big Confusion: `self` Kya Hota Hai?

Aksar students `self` par aakar phans jate hain.

**Explanation:** `self` ka matlab hai **"Main Khud"** (The current object). Jab `student1.introduce()` call hota hai, toh Python ko kaise pata chalega ke "Zohaib" ka naam print karna hai ya "Ali" ka?

**Inner Working:** Background mein Python is call ko aise run karta hai:
`Student.introduce(student1)`

Yahan `student1` as a parameter pass hota hai aur method ke andar `self` ban jata hai. Agar aap `self` nahi likhenge, toh object ko pata hi nahi chalega ke kis data par kaam karna hai.

---

## 4. Class Variables vs. Instance Variables

Yeh concept memory aur data management samjhane ke liye zaroori hai.

```python
class Employee:
    # Class Variable (Sab objects ke liye same rehta hai)
    company_name = "Tech Solutions"

    def __init__(self, emp_name):
        # Instance Variable (Har object ka apna alag hota hai)
        self.emp_name = emp_name

emp1 = Employee("Ahmed")
emp2 = Employee("Sara")

print(emp1.company_name) # Output: Tech Solutions
print(emp2.company_name) # Output: Tech Solutions
```


* **Instance Variable (`self.name`):** Har student/employee ka apna personal data.
* **Class Variable (`company_name`):** Shared data. Agar class variable change hoga, toh sab objects ke liye change ho jayega.

## 5. Inner Workings: Objects Asal Mein Kya Hain? (Deep Dive)

Jab aapki class advanced level ka soche, toh unhein batayein ke **Python mein har cheez ek Dictionary hai.**

Objects ka sara data actually ek dictionary mein save ho raha hota hai jise `__dict__` kehte hain.


```python
# Agar hum purane student1 ko check karein
print(student1.__dict__)
# Output: {'name': 'Zohaib', 'roll_number': '101'}
```


Jab hum `student1.name` call karte hain, Python actually is `__dict__` mein ja kar 'name' key ki value dhoondhta hai.

Iske ilawa, Python mein Classes khud bhi Objects hain (jinhein *Metaclasses* banati hain). Isiliye aap class ko bhi kisi function mein as an argument pass kar sakte hain.

## 6. Dunder / Magic Methods (`__str__`, `__len__`)

Python mein double underscores `__` wale methods ko Dunder methods kehte hain. Yeh class ke behavior ko customize karte hain.

Agar aap direct `print(student1)` karenge, toh Python memory location print karega (e.g., `<__main__.Student object at 0x00...>`). Isko fix karne ke liye `__str__` use hota hai:

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
        
    def __str__(self):
        # Yeh define karta hai ke object ko string (print) mein kaise dikhana hai
        return f"Book: {self.title}"
        
    def __len__(self):
        # Is se hum object par len() use kar sakte hain
        return self.pages

my_book = Book("Python Mastery", 300)
print(my_book)        # Output: Book: Python Mastery
print(len(my_book))   # Output: 300
```


## 7. Aakhri Concept: Inheritance (Virasat)

Code reusability samjhane ke liye. Ek parent class ki khususiyaat (attributes/methods) child class mein lana.

```python
# Parent Class
class Animal:
    def eat(self):
        print("I can eat")

# Child Class
class Dog(Animal):
    def bark(self):
        print("Woof!")

my_dog = Dog()
my_dog.eat()  # Inherited from Animal
my_dog.bark() # Dog's own method
```
