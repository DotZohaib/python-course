# 1. `for` Loop (The Sequence Reader)

**Proper Definition:**
`for` loop ek aisi statement hai jo kisi bhi Data Structure (jaise sequence, list, ya string) ke andar maujood har ek item par baari-baari (step-by-step) jata hai aur us par koi action perform karta hai. Isko pehle se pata hota hai ke data kitna lamba hai, is liye yeh khud ba khud wahan ja kar ruk jata hai.

## Example A: Ek String (Lafz) ko todna

Jab hum chahte hain ke kisi lafz ke har letter par loop chale.

```python
word = "PYTHON"

for letter in word:
    print(f"Current Letter is: {letter}")
```

**Inner Working:**

1. Python check karega ke "PYTHON" mein kitne letters hain (Total 6).
2. Pehle chakkar mein woh pehla letter **'P'** uthayega aur usey `letter` naam ke variable mein daal dega, phir `print` ki line chalayega.
3. Dusre chakkar mein woh khud **'Y'** par chala jayega.
4. Yeh process tab tak chalega jab tak aakhri letter **'N'** print nahi ho jata.

## Example B: List of Students (Real-World Data)


```python
students = ["Ali", "Sara", "Ahmed", "Zainab"]

for name in students:
    print(f"Welcome to the class, {name}!")
```

**Inner Working:**
Loop list ke andar maujood items ko index (position) ke hisaab se parhta hai. Woh pehle position 0 ("Ali") ko uthata hai, apna kaam karta hai, phir position 1 ("Sara") par jata hai. Hamein loop ko batana nahi parta ke agle naam par jao, woh khud jata hai.


```python
for i in range(3,6):
    print(i)
```

```python
# 1. Starting Point
count = 1 

# 2. Condition (Jab tak count 5 ya us se chota hai)
while count < 5: 
    print("Number hai:", count)
    
    # 3. Step (Har dafa count mein 1 add karo)
    count = count + 1 
    
print("Loop khatam ho gaya!")
```


1. **Shuruaat (Initialization):** Line: count = 1.
Computer apni memory (dimagh) mein ek dabbi (box) banata hai jiska naam woh `count` rakhta hai, aur uske andar **1** daal deta hai.


2. **Pehla Chakkar (Iteration 1):** count ki value ab 1 hai.
* **Shart Check:** Kya `1 <= 5` hai? (Haan, 1 chota hota hai 5 se, True).
* **Andar aao:** Screen par print karega: `Number hai: 1`
* **Step:** `count = 1 + 1` karega. Ab `count` ke dabbe mein **2** aa gaya. Loop wapas upar shart check karne jayega.


3. **Doosra Chakkar (Iteration 2):** count ki value ab 2 hai.
* **Shart Check:** Kya `2 <= 5` hai? (Haan, True).
* **Andar aao:** Screen par print karega: `Number hai: 2`
* **Step:** `count = 2 + 1` karega. Ab `count` ki value **3** ho gayi. Loop phir upar jayega.


4. **Teesra aur Chautha Chakkar:** Yahi amal doharaya jayega.
Jab count **3** aur phir **4** hoga, tab bhi shart sach hogi, aur screen par numbers print hote rahenge. Unke baad `count` ki value mein 1 ka izafa ho kar aakhir kaar **5** ho jayegi.


5. **Panchwan Chakkar (Iteration 5):** count ki value ab 5 hai.
* **Shart Check:** Kya `5 <= 5` hai? (Haan, kyunki humne `<=` "barabar" bhi lagaya hai, True).
* **Andar aao:** Screen par print karega: `Number hai: 5`
* **Step:** `count = 5 + 1` karega. Ab `count` ke dabbe mein **6** aa gaya. Loop wapas aakhri dafa upar jayega.


6. **Loop Toot Gaya (Condition False):** count ki value ab 6 hai.
* **Shart Check:** Kya `6 <= 5` hai? (**Nahi!** 6 na toh 5 se chota hai aur na barabar, yeh **False** ho gaya).
* Jaise hi condition False hui, computer loop ke andar jana band kar dega aur us block ko chhod dega.


7. **Loop Ke Bahar Ka Kaam:** Line: print('Loop khatam ho gaya!').
Computer seedha while loop ki boundary se baahar aayega aur aakhri line chalayega. Screen par aayega: `Loop khatam ho gaya!`


## 3. Nested Loops (Loop ke andar Loop)

**Concept:** Ek main kaam ke andar chote chote kaam baar baar karna. Sab se behtareen misaal **Ghari (Clock)** ki hai. Ghari ka ghanta (hour) 1 dafa badalta hai, jabke minute wali sui us ek ghante ke andar poore 60 chakkar lagati hai.

### Example D: The Clock Logic


```python
# Outer Loop: Yeh ghanton (Hours) ke liye hai
for hour in range(1, 25):  # 1 se le kar 24 bajne tak
    print(f"\n--- Hour {hour} starts ---")
    
    # Inner Loop: Yeh Minutes ke liye hai
    for minute in range(1, 61):  # Hum sirf 3 minute chala kar dikhayenge
        print(f"  Minute {minute}")
```


**Inner Working:**

1. Pehle **Outer Loop** chalta hai. `hour` ki value `1` hoti hai.
2. Code neeche aata hai aur **Inner Loop** mein enter ho jata hai.
3. Ab Outer loop tab tak aage nahi barh sakta, jab tak Inner loop apne saare chakkar (1, 2, 3 minute) mukammal na kar le.
4. Jab Inner loop khatam hota hai, tab engine wapas Outer loop ke paas jata hai aur `hour` ki value `2` karta hai, aur phir wapas naye sire se Inner loop ko 3 dafa chalata hai.

```python
# range(1, 6) ka matlab hai 1 se shuru karo aur 6 se pehle (5 par) ruk jao.
for i in range(1, 6):
    print(f"Chakkar number: {i}")
```


### Inner Working (Peeche kya ho raha hai?):

1. **Sequence Ready Hona:** Python apne dimaagh mein ek sequence banata hai (jaise `1, 2, 3, 4, 5`).
2. **Variable Assignment:** Loop sequence ka pehla number uthata hai aur usey variable (`i`) mein daal deta hai. Pehli dafa `i` ki value 1 hoti hai.
3. **Execution:** Loop ke andar ka code chalta hai aur screen par print hota hai.
4. **Auto-Increment:** Jab andar ka code poora ho jata hai, loop khud-ba-khud sequence ke agle number (2) ko `i` mein daalta hai. Hamein khud +1 nahi karna parta.
5. **Termination:** Jaise hi sequence ke items (5 tak) khatam hote hain, loop automatically ruk jata hai.


## 3. `while` Loop (Condition-Based Loop)

**Concept:** `while` loop tab use hota hai jab aapko **yeh na pata ho ke kitne chakkar lagenge**, bas ek shart (condition) pata ho. Iska asool hai: "Jab tak meri condition sach (True) hai, tab tak code chalate raho."


```python
battery = 3

while battery > 0:
    print(f"Game chal rahi hai... Battery bachi hai: {battery}")
    battery = battery - 1  # Battery ko har dafa kam karna zaroori hai!
    
print("Battery khatam, Game Over!")
```

## Inner Working (Peeche kya ho raha hai?):

1. **Condition Check:** Loop chalne se pehle `while` ki condition (`battery > 0`) check hoti hai.
2. **Execution:** Agar condition **True** hai (jaise 3 > 0), toh andar ka block chalta hai.
3. **The Loop Back:** Block ki aakhri line chalne ke baad, engine wapas **upar** line 1 par jata hai aur dobara condition check karta hai.
4. **Manual Update (Important!):** Isme value khud-ba-khud update nahi hoti. Aapko andar khud likhna padta hai (`battery = battery - 1`), warna condition hamesha True rahegi aur computer **Infinite Loop** (kabhi na rukne wala loop) mein phans jayega aur program crash ho jayega.


## "Summary Formula"

Students ko in dono ka farq hamesha ke liye yaad karwane ke liye yeh misaal dein:

* **Known Steps = `for` Loop:** Agar track ki lambai pata hai (jaise "Mujhe theek 10 round lagane hain"), toh `for` loop lagao.
* **Unknown Steps + Condition = `while` Loop:** Agar track ki lambai nahi pata, bas shart yeh hai ke "Jab tak paseena na aa jaye tab tak bhagna hai", toh `while` loop lagao.

## Ek Zaroori Tool: `range()` Function

Python mein `for` loop `range()` ke bina adhoora hai. Students ko iske 3 parameters ka magic zaroor samjhayein:
`range(start, stop, step)`

* `start`: Kahan se ginti shuru karni hai.
* `stop`: Kahan rukna hai (yeh number shamil nahi hota, ek pehle ruk jata hai).
* `step`: Kitne numbers ka jump lagana hai.

```python
# 2 se shuru hoga, 11 se pehle rukega, aur 2-2 ka jump lega (Even numbers)
for num in range(2, 11, 2):
    print(num)  # Output aayega: 2, 4, 6, 8, 10
```

 `while` Loop (The Condition Checker)

**Proper Definition:**
`while` loop ek aisi statement hai jo us waqt tak apna code lagataar chalati rehti hai jab tak usko di gayi shart (condition) **True (sach)** rehti hai. Jaise hi woh shart **False (jhoot)** hoti hai, loop wahi ruk jata hai aur program aage barh jata hai.


# Example C: ATM PIN ka logic (Real-World Login)

Aap ATM mein apna card dalte hain. Jab tak aap galat PIN daalte rahenge, ATM aap se baar baar PIN mangta rahega.

```python
correct_pin = "1234"
entered_pin = ""  # Shuru mein yeh khali hai

# Jab tak entered_pin aur correct_pin aapas mein barabar NAHI hain (!=)
while entered_pin != correct_pin:
    entered_pin = input("Please enter your 4-digit PIN: ")
    
print("Login Successful! Welcome to your account.")
```

**Inner Working:**

1. Pehle step par condition check hoti hai: Kya khali string `""` barabar nahi hai `"1234"` ke? Jawab hai: Haan, barabar nahi hai (True).
2. Kyunki condition True hai, loop andar jayega aur user se input mangega.
3. Agar user ne `"0000"` likha, loop wapas upar jayega. Check karega: Kya `"0000"` barabar nahi hai `"1234"` ke? True! Loop phir se input mangega.
4. Agar user ne `"1234"` likha, loop wapas upar jayega. Kya `"1234"` barabar nahi hai `"1234"` ke? Jawab hai: False (Kyunki ab barabar ho gaye hain).
5. Condition False hote hi loop foran destroy ho jayega aur "Login Successful!" print ho jayega.
