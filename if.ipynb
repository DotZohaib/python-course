# 1. `if`, `elif`, aur `else` (Faisla Karna)

**Asaan Alfaaz Mein (Concept):**
Zindagi mein hum har roz faisle karte hain. "Agar (if) barish hui, toh chata (umbrella) le lo, warna (else) na lo." Programming mein bhi computer ko decision lene ke liye `if-else` ki zaroorat hoti hai.

```python
marks = 75

if marks >= 80:
    print("A Grade")
elif marks >= 50:
    print("Pass")
else:
    print("Fail")
```

# Inner Working (Peeche kya ho raha hai?):

1. **Line by Line Execution:** Computer code ko upar se neeche parhta hai.
2. **Condition Checking:** Sab se pehle woh `if` ki condition (`marks >= 80`) check karega.
3. **True / False Game:**
* Agar condition **True** hai, toh woh uske andar ka code chalayega aur baaki sab (`elif`, `else`) ko mukammal taur par ignore kar dega.
* Agar condition **False** hai, tabhi woh next `elif` par jayega.


4. **Default Catch:** Agar koi bhi condition True nahi hoti, toh aakhir mein `else` lazmi chalta hai (yeh ek backup plan hota hai).



# 2. `break` Statement (Emergency Brake)

**Asaan Alfaaz Mein (Concept):**
Sochiye aap class mein 50 bacchon mein se 'Ali' ko dhoondh rahe hain. Jaise hi 5th number par Ali mil gaya, kya aap baaki 45 bacchon ka naam pukarenge? Nahi! Aap wahi ruk jayenge. `break` loop ka emergency brake hai.

```python
for number in range(1, 10):  # Loop 1 se 9 tak chalega
    if number == 5:
        print("Brake lag gaya!")
        break  # Loop yahi khatam
    print(number)
```

# Inner Working (Peeche kya ho raha hai?):

1. Loop apna normal chakkar (iteration) laga raha hota hai.
2. Jaise hi program `break` word ko parhta hai, woh **loop ko foran destroy (terminate)** kar deta hai.
3. Chahe loop ke 100 chakkar baaki hon, control foran loop se bahar aa jata hai aur loop ke aage ka code chalne lagta hai.
4. **Rule of thumb:** `break` hamesha kisi na kisi condition (`if`) ke andar lagta hai, warna loop pehli dafa mein hi ruk jayega.


# 3. `continue` Statement (Sirf Ek Turn Skip Karna)

**Asaan Alfaaz Mein (Concept):**
Maan lijiye aap sab bacchon ko chocolates baant rahe hain, lekin ek baccha absent hai. Aap kya karenge? Us bacchay ko skip karenge aur agle bacchay ko chocolate de denge. Aap chocolate baantna (loop) rok thodi denge! `continue` sirf **current step ko skip** karta hai.


```python
for number in range(1, 6):  # 1 se 5 tak chalega
    if number == 3:
        print("3 ko skip kar diya")
        continue  # Neeche wala code nahi chalega, loop wapas upar chala jayega
    
    print(number)
```

# Inner Working (Peeche kya ho raha hai?):

1. Jaise hi program `continue` ko dekhta hai, uske neeche likha hua loop ka koi bhi code us turn ke liye **execute nahi hota**.
2. Loop break *nahi* hota, balki engine foran wapas upar loop ke start mein chala jata hai agla number (next iteration) lene ke liye.
3. Pura loop complete hota hai, sirf wo specific condition skip ho jati hai.

# Master Example

```python
student_ids = [101, 102, 103, 104, 105]
suspended_student = 103
target_student = 104

for student in student_ids:
    if student == suspended_student:
        print(f"Student {student} suspended hai, skip karo.")
        continue  # Isko skip kar ke agle par jao
        
    if student == target_student:
        print(f"Target mil gaya: Student {student}. Searching rok do.")
        break  # Target mil gaya, loop khatam karo
        
    print(f"Checking Student: {student}")
```


**Tip:** humesha Yaad rakhin ke `break` aur `continue` bina `if` condition ke use karna be-maqsad hai. Yeh hamesha kisi khaas situation ko handle karne ke liye banaye gaye hain.
