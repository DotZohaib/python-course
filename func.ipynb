Programming ka ek asool hai **DRY (Don't Repeat Yourself)**. Agar aapko ek hi kaam 10 jagah karna hai, toh 10 dafa code likhne ke bajaye ek function bana lein aur usey 10 dafa pukar lein.

# 2. Function Banate Kaise Hain? (Syntax & Basic Example)

Python mein function banane ke liye hum `def` (Define) keyword ka istemaal karte hain.

```python
# 1. Function banana (Machine tayyar karna)
def say_hello():
    print("Hello! Welcome to the class.")
    print("Umeed hai aap theek honge.")

# 2. Function ko call karna (Machine ka button dabana)
say_hello()
```


## Inner Working (Peeche kya ho raha hai?):

1. **Definition Phase:** Jab computer `def say_hello():` wali line parhta hai, toh woh andar ka code **chalata nahi hai**. Woh bas us code ko apni memory (dimaagh) mein ek box mein band kar ke us box ka naam `say_hello` rakh deta hai.
2. **Calling Phase:** Jab computer line number 7 par `say_hello()` dekhta hai, toh uska main engine wahan ruk jata hai.
3. Woh dimaagh mein ja kar `say_hello` wala box kholta hai, uske andar ki saari lines (print statements) chalata hai, aur jab box khatam ho jata hai, toh wapas usi jagah aa jata hai jahan se gaya tha.

---

## 3. Parameters aur Arguments (Function ko Input Dena)

Function tabhi mazedar hota hai jab hum usko apni marzi ka data dein aur woh us par kaam kare.

* **Parameter:** Function banate waqt jo variable hum bracket `()` mein likhte hain (Machine ka sancha).
* **Argument:** Function call karte waqt jo asli data hum usko dete hain (Asli fruit).

```python
# 'name' ek parameter hai
def greet_student(name):
    print(f"Welcome to the Python class, {name}!")

# "Ali" aur "Sara" arguments hain
greet_student("Ali")
greet_student("Sara")
```

## Inner Working:

Jab hum `greet_student("Ali")` likhte hain, toh Python chupke se `name = "Ali"` kar deta hai aur phir andar ka code chalata hai. Agli dafa woh `name = "Sara"` kar deta hai. Ek hi code alag alag data par chal raha hai!

---

## 4. `return` Statement (Function se Jawab Wapas Lena)

**Class ke liye zaroori point:** `print()` aur `return` mein zameen aasman ka farq hai.

* `print()` sirf screen par dikhata hai.
* `return` asli data wapas bhejta hai taake hum usko kisi variable mein save kar sakein ya mazeed calculation mein use kar sakein. (Jaise Juicer machine juice wapas deti hai, sirf juice ki tasveer nahi dikhati).

```python
def add_numbers(num1, num2):
    total = num1 + num2
    return total  # Answer wapas bhej raha hai

# Function call kiya aur jo jawab aaya usey 'result' mein save kar liya
result = add_numbers(50, 30)

print(f"Total marks are: {result}")
```


## Inner Working:

1. `add_numbers(50, 30)` call hua.
2. Python ne `num1` ko 50 aur `num2` ko 30 diya.
3. Dono ko plus kar ke `total` (80) banaya.
4. `return 80` ne chup-chaap 80 ko wapas wahan bhej diya jahan se function call hua tha. Ab `result` variable ke andar 80 save ho gaya hai. Jaise hi function `return` karta hai, woh foran wahi par destroy (ruk) jata hai.

## 5. Default Parameters (Back-up Option)

Kabhi kabhi hum chahte hain ke agar user koi input na de, toh function error dene ke bajaye ek default (pehle se set) value utha le.

```python
# Agar koi country nahi batayega, toh function khud "Pakistan" maan lega
def show_profile(name, country="Pakistan"):
    print(f"Student: {name}, Country: {country}")

show_profile("Ahmed", "UAE")  # Yahan UAE print hoga
show_profile("Zainab")        # Yahan koi country nahi di, toh default "Pakistan" print hoga
```

## 6. Types of Functions in Python

Class ko clear karne ke liye batayein ke Python mein 2 tarah ke functions hote hain:

| Function Type | Explanation | Examples |
| --- | --- | --- |
| **Built-in Functions** | Yeh woh functions hain jo Python walon ne pehle se bana kar rakhe hain. Hamein bas inka naam pukar kar kaam nikalwana hota hai. | `print()`, `len()`, `type()`, `range()` |
| **User-Defined Functions** | Yeh woh functions hain jo hum (programmers) apni zaroorat ke hisaab se khud banate hain `def` ka word use kar ke. | `add_numbers()`, `say_hello()` |


```python
def check_grade(marks):
    # Function ka kaam sirf grade calculate kar ke wapas dena hai
    if marks >= 80:
        return "A Grade"
    elif marks >= 60:
        return "B Grade"
    else:
        return "Fail"

# Main Program
student_marks = [85, 45, 72]

for marks in student_marks:
    # Hum loop ke andar apne banaye hue function ko call kar rahe hain
    grade = check_grade(marks)

    print(f"Marks: {marks} | Result: {grade}")
```
