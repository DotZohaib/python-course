# 2. Variables Kya Hain aur Kaise Bante Hain?

**Concept (Analogy):**
"Variable ko aap ek **'Dibba' (Box) ya 'Container'** samajh lein. Jaise aapke kitchen mein cheeni (sugar) rakhne ke liye ek jar hota hai aur us par 'Sugar' ka label laga hota hai. Yahan 'Sugar' variable ka naam hai, aur uske andar rakhi cheeni uski 'Value' (Data) hai."

Programming mein jab humein koi data (naam, age, marks) computer ki memory mein save karna hota hai, toh hum variable banate hain.

# Variable Banane Ka Tarika (How to create):

Python mein variable banana bohat asaan hai. Aapko koi special word likhne ki zaroorat nahi hoti. Bas naam likhein, `=` lagayen, aur value de dein.

**Rules:**

* Variable ke naam mein spaces ( ) nahi aa sakti.
* Variable ka naam kisi number (1, 2, 3) se start nahi ho sakta.

```python
# Text (String) data store karna
student_name = "Ali"   

# Number (Integer) data store karna
age = 20               

# Point wala number (Float) store karna
marks = 85.5           

# Variables ko screen par dikhana
print(student_name)
print(age)
print(marks)
```

# 3. Operators Kya Hain?

**Concept:**
"Operators wo special symbols (`+`, `-`, `=`, `>`) hote hain, jo variables aur values ke darmiyan koi operation (kaam) perform karte hain. Jaise math mein 2 aur 2 ko jorne ke liye hum `+` ka sign use karte hain."

Python mein operators ki 4 main types hain jo students ko shuru mein sikhani zaroori hain:

# Type 1: Arithmetic Operators (Math ke kamo ke liye)

Yeh mathematical calculation ke liye use hote hain.

* `+` (Plus)
* `-` (Minus)
* `*` (Multiply)
* `/` (Divide)
* `%` (Modulus - Remainder ya bachi hui value nikalne ke liye)

```python
a = 10
b = 3

print(a + b)  # Addition, Output: 13
print(a - b)  # Subtraction, Output: 7
print(a * b)  # Multiplication, Output: 30
print(a / b)  # Division, Output: 3.333
print(a % b)  # Modulus (10 ko 3 se divide karein toh 1 bachta hai), Output: 1
```

# Type 2: Assignment Operators (Value assign karne ke liye)

Yeh variables ke andar value dalne ke liye use hote hain.

* `=` (Direct value dalna)
* `+=` (Pehle wali value mein mazeed add kar ke save karna)

```python
score = 50      # '=' operator ne 50 ko score mein daal diya
score += 10     # Iska matlab hai: score = score + 10 (Ab score 60 ho gaya)

print(score)    # Output: 60
```

# Type 3: Comparison Operators (Muazna / Faisla karne ke liye)

Yeh do values ko compare karte hain. Inka answer hamesha **True** (Sahi) ya **False** (Ghalat) mein aata hai.

* `==` (Kya dono barabar hain?)
* `!=` (Kya dono barabar NAHI hain?)
* `>` (Kya pehli value bari hai?)
* `<` (Kya pehli value choti hai?)

```python
x = 10
y = 20

print(x == y)  # Kya x aur y barabar hain? Output: False
print(x < y)   # Kya x chota hai y se? Output: True
print(x != y)  # Kya x aur y barabar nahi hain? Output: True
```

# Type 4: Logical Operators (Conditions check karne ke liye)

Yeh check karte hain ke ek se zyada conditions sahi hain ya nahi.

* `and` (Dono conditions ka True hona zaroori hai)
* `or` (Dono mein se koi ek bhi True ho toh chalega)
* `not` (True ko False aur False ko True kar deta hai)

```python
# Real life example: University aane ke liye ID Card OR Fees slip mein se ek lazmi hai
has_id_card = False
has_fees_slip = True

# 'or' check karega kya dono mein se ek cheez hai?
print(has_id_card or has_fees_slip)  # Output: True (Student andar ja sakta hai)

# 'and' check karega kya dono cheezein hain?
print(has_id_card and has_fees_slip) # Output: False (Dono nahi hain)

# 'not' operator ka matlab hai ulta karna
print(not has_id_card)  # Output: True (kyunki has_id_card False hai
```
